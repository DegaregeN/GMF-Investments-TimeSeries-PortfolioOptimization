```python
import yfinance as yf
import pandas as pd
```

```python
def download_data(ticker, start_date='2015-01-01', end_date='2024-10-31'):
    """Download historical financial data from YFinance."""
    data = yf.download(ticker, start=start_date, end=end_date)
    data['Ticker'] = ticker
    return data

def main():
    assets = ['TSLA', 'BND', 'SPY']
    all_data = []

    for ticker in assets:
        data = download_data(ticker)
        data.to_csv(f'../data/raw/{ticker}_data.csv')  # Save raw data

if __name__ == "__main__":
    main()
```

## Reading of TSLA data

```python
# Load Tesla dataset
df_tesla = pd.read_csv('../data/raw/TSLA_data.csv')
```

```python
# Dropping 1st two rows by index
indexes_to_drop = [0, 1]  # Specify the row indices to drop
df_tesla = df_tesla.drop(indexes_to_drop)
```

```python
# Renaming a Price with date
df_tesla = df_tesla.rename(columns={'Price': 'Date'})
df_tesla
```
